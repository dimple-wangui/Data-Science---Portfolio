```python
# --- STEP 1: Mount Google Drive ---
from google.colab import drive
drive.mount('/content/drive')

# --- STEP 2: Import libraries ---
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime

# --- STEP 3: Define target URL to be scrapped ---
url = "https://explodingtopics.com/blog/data-generated-per-day"
headers = {"User-Agent": "Mozilla/5.0"}

# --- STEP 4: Fetch and parse HTML ---
response = requests.get(url, headers=headers)
response.raise_for_status()
soup = BeautifulSoup(response.text, "html.parser")

# --- STEP 5: Extract data ---
data_stats = []

# Find all tables on the page
all_tables = soup.find_all("table")

target_table = all_tables[1]

for row in target_table.find_all("tr")[1:]:  # [1:] skips the header row
    cols = row.find_all(["th", "td"])

    #Labelling the columns according to the lecture notes table
    if len(cols) == 4:
        data_stats.append({
            "Year": cols[0].get_text(strip=True),
            "Data Generated": cols[1].get_text(strip=True),
            "Change Over Previous Year": cols[2].get_text(strip=True),
            "Change Over Previous Year (%)": cols[3].get_text(strip=True)
        })

# --- STEP 6: Save data to CSV ---
today = datetime.now().strftime("%Y-%m-%d")

drive_path = f"/content/drive/MyDrive/zettabytes_growth_{today}.csv"

with open(drive_path, "w", newline="", encoding="utf-8") as f:

    fieldnames = [
        "Year",
        "Data Generated",
        "Change Over Previous Year",
        "Change Over Previous Year (%)"
    ]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data_stats)

    print(f"Success! Data saved to Google Drive:\n{drive_path}")
```
